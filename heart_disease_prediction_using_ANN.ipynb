```python
#importing the required libraires
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import keras
from keras.models import Sequential
from keras.layers import Dense
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
```

```python
#importing and accessing the dataset
df = pd.read_csv("heart.csv")
df.head()
```

```python
#creating a summary statistics
df.describe()
```

```python
#checking whether there are any missing values
df.isnull().any()
```

```python
#extracting the features (X) and the target variable (y) from the dataframe (df)
X = df.iloc[:,:13].values
y = df["target"].values
```

```python
#splitting the data into training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.3 , random_state = 0 )
```

```python
#standardizing the features in the training and testing datasets
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
#defining a neural network model using keras Sequential API
classifier = Sequential()
classifier.add(Dense(activation = "relu", input_dim = 13,
					units = 8, kernel_initializer = "uniform"))
classifier.add(Dense(activation = "relu", units = 14,
					kernel_initializer = "uniform"))
classifier.add(Dense(activation = "sigmoid", units = 1,
					kernel_initializer = "uniform"))
classifier.compile(optimizer = 'adam' , loss = 'binary_crossentropy',
				metrics = ['accuracy'] )
```

```python
  #training the neural network model using fit method in keras
  classifier.fit(X_train , y_train , batch_size = 8 ,epochs = 100 )
```

```python
#making predictions with the trained neural network model and thresholding those predictions to make them binary (either 0 or 1) based on a cutoff of 0.5
y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5)
```

```python
#calculating and displaying confusion matrix for model's prediction
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
#calculating accuracy of the model
accuracy = (cm[0][0]+cm[1][1])/(cm[0][1] + cm[1][0] +cm[0][0] +cm[1][1])
print(accuracy*100)
```
